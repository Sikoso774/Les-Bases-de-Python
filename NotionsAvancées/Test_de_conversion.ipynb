````python
import base64

def convertir_image_en_base64(chemin_image):
    """
    Lit une image √† partir de son chemin d'acc√®s et la convertit en une cha√Æne Base64.

    Args:
        chemin_image (str): Le chemin d'acc√®s au fichier image.

    Returns:
        str: La cha√Æne Base64 de l'image, ou None en cas d'erreur.
    """
    try:
        with open(chemin_image, "rb") as image_file:
            encoded_string = base64.b64encode(image_file.read()).decode('utf-8')
        return encoded_string
    except FileNotFoundError:
        print(f"Erreur : Le fichier image '{chemin_image}' n'a pas √©t√© trouv√©.")
        return None
    except Exception as e:
        print(f"Erreur lors de la conversion de l'image en Base64 : {e}")
        return None

# Instructions d'utilisation :
# 1. Assurez-vous d'avoir l'image que vous souhaitez convertir enregistr√©e sur votre ordinateur.
# 2. Remplacez 'chemin_vers_votre_image.png' (ou l'extension appropri√©e) par le chemin r√©el de votre fichier image.

chemin_de_l_image = 'Image_test.png'
base64_data = convertir_image_en_base64(chemin_de_l_image)

if base64_data:
    print("Voici la cha√Æne Base64 de votre image :")
    print(base64_data)
    print("\n")
    print("Vous pouvez utiliser cette cha√Æne dans votre Markdown comme ceci :")
    print("```markdown")
    # Pour une image PNG
    print(f"![Votre description de l'image](data:image/png;base64,{base64_data})")
    # Si votre image est un JPEG, remplacez 'png' par 'jpeg'
    # print(f"![Votre description de l'image](data:image/jpeg;base64,{base64_data})")
    print("```")
````
